#Funciones




```python
def greet():
    print("Hello, world!")

greet()
```

```python
def show_name(name):
    print("Hello", name)

show_name("Enith")
```

```python
def add(a, b):
    return a + b

print(add(3, 5))
```

```python
def subtract(a, b):
    return a - b

print(subtract(10, 4))
```

```python
def multiply(a, b):
    return a * b

print(multiply(3, 6))
```

```python
def divide(a, b):
    return a / b

print(divide(10, 2))
```

```python
def is_even(number):
    return number % 2 == 0

print(is_even(4))
```

```python
def find_max(a, b):
    return max(a, b)

print(find_max(7, 10))
```

```python
def square(number):
    return number ** 2

print(square(5))
```

```python
def cube(number):
    return number ** 3

print(cube(3))
```

```python
def count_letters(word):
    return len(word)

print(count_letters("python"))
```

```python
def convert_to_upper(text):
    return text.upper()

print(convert_to_upper("hola"))
```

```python
def calculate_area(length, width):
    return length * width

print(calculate_area(5, 3))
```

```python
def average(numbers):
    return sum(numbers) / len(numbers)

print(average([1, 2, 3, 4, 5]))
```

```python
def factorial(n):
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result

print(factorial(5))
```
